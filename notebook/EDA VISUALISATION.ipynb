```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
```

```python
data = pd.read_csv('data/data.csv')
```

## No of Male and Female Candidates


```python
sns.countplot(data["SEX"])
plt.title("Female v/s Male")
plt.savefig('figure1.jpeg')
```

```python
sex_education = (data[['SEX','EDUCATION']].query("SEX=='Female' |SEX == 'Male' ").groupby(['EDUCATION']).value_counts())
```

```python
sex_education.unstack(level=0).plot(kind='bar')
plt.title("Number of Male and Female customer with repect to Education")
plt.savefig('figure2.jpeg')
```

```python
values = [data['MARRIAGE'].value_counts()[i] for i in range(len(data['MARRIAGE'].value_counts()))]
labels = ['Single','Married','Other']

colors = sns.color_palette('pastel')[0:3]
plt.pie(values, labels = labels, colors = colors, autopct='%.0f%%')
plt.title("Marital status v/s Output")
plt.savefig("figure3.jpeg")
```

```python
marriage_education = data[['MARRIAGE','EDUCATION']].groupby(['EDUCATION','MARRIAGE']).value_counts(ascending=True)
```

```python
marriage_education.unstack(level=0).plot(kind='barh')
plt.title("Relation between customer Marital status and Education")
plt.savefig('figure4.jpeg')
```

```python
Education_vs_Output = data[['default#payment#next#month','EDUCATION']].groupby(['EDUCATION','default#payment#next#month']).value_counts(ascending=True)
Education_vs_Output.unstack(level=0).plot(kind="bar")
plt.title("Relation between Customer's Education and Output")
plt.savefig('figure5.jpeg')
```
