## Outlier treatment for Bill_amt coulmns and Pay columns

```python
def outlier_removal(feature_name,quantile_val):
    data.loc[data[feature_name]>=data[feature_name].quantile(quantile_val),feature_name] = data[feature_name].quantile(quantile_val)  
    
    return data[feature_name]
```

```python
data[["BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6" ]].describe(percentiles=[0.75,0.80,0.85,0.90,0.95])
```

```python
bill_column_list = ["BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6" ]
for i in bill_column_list:
    data[i] = outlier_removal(i,0.90)
```

```python
data[["BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6" ]].describe(percentiles=[0.75,0.80,0.85,0.90,0.95])
```

```python
data[['PAY_AMT1','PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']].describe(percentiles=[0.75,0.80,0.85,0.90,0.95])
```

```python
pay_column_list = ['PAY_AMT1','PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
for i in pay_column_list:
    data[i] = outlier_removal(i,0.90)
```

## Negative value treatment

```python

def negative_val_treatment(feature_name):
    data.loc[data[feature_name]<=0,feature_name] = abs(data.loc[data[feature_name]<=0,feature_name])
    return data[feature_name]
```

```python
pay_column_list = ['PAY_AMT1','PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
for i in pay_column_list:
    data[i] = negative_val_treatment(i)
```

```python
bill_column_list = ["BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6"]
for i in bill_column_list:
    data[i] = negative_val_treatment(i)
```
